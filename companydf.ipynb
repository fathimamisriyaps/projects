### Data Cleaning

In the Dataset we can drop following columns :

```python
company_df.drop(['Country','State lgd code','Corporate identification number (cin)','Company name',
                'Company roc code','Company registration number','Company registration date','Registered office address'],axis=1,inplace=True)
```

```python
company_df.head()
```

```python
#handling missing values

company_df['Listing status']=company_df['Listing status'].replace('Not Available',np.NaN)
company_df.isna().sum()
```

```python
#fillna()
company_df['Listing status']=company_df['Listing status'].fillna(company_df['Listing status'].mode()[0])
company_df.isna().sum()
```

### removing outliers

```python
company_df.plot(kind='box',figsize=(12,12),sharex=False,subplots=True)
```

```python
outliers_cols=['YearCode','Authorized capital','Paidup capital']
for column in outliers_cols:
    if company_df[column].dtype in ["int64","float64"]:
        q1 = company_df[column].quantile(0.25)
        q3 = company_df[column].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        company_df = company_df[(company_df[column] >= lower_bound) & (company_df[column] <= upper_bound)]
company_df
```

```python
company_df.plot(kind='box',figsize=(12,12),sharex=False,subplots=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encod=LabelEncoder()
encod_cols=company_df.select_dtypes(include=['object']).columns
for i in encod_cols:
  company_df[i]=encod.fit_transform(company_df[i])
company_df.head()
```

```python
corr=company_df.corr()
sns.heatmap(corr)
```

```python
x=company_df.iloc[:,:]
x
```

```python
#finding number of clusters
#graph ,varience
#elbow point
from sklearn.cluster import KMeans
wcss=[]                                       #empty list
for i in range(1,11):                         # i=1 to 10
  data=KMeans(n_clusters=i,init='k-means++',random_state=42)
  data.fit(x)
  wcss.append(data.inertia_)   #to find varience
```

```python
print(wcss)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.xlabel('number of clusters')
plt.ylabel('varience')
plt.title('elbow point evaluation')
```

```python

model=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_means=model.fit_predict(x)
y_means
```

```python
company_df['output']=y_means
company_df
```

```python
x1=company_df.iloc[:,:-1].values
x1
```

```python
y=company_df.iloc[:,-1].values
y
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
knn=KNeighborsClassifier()
gnb=GaussianNB()
sv=SVC()
```

```python
kvalue={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
gd=GridSearchCV(knn,kvalue,cv=10,scoring='accuracy')
gd.fit(x_train,y_train)
```

```python
print(gd.best_params_)
```

```python
model=DecisionTreeClassifier(criterion='entropy')
k_model=KNeighborsClassifier(n_neighbors=9,weights='uniform')
n_model=GaussianNB()
s_model=SVC()
lst_model=[k_model,n_model,s_model]
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay
for i in lst_model:
   print('model name',i)
   i.fit(x_train,y_train)
   y_pred=i.predict(x_test)
   score=accuracy_score(y_test,y_pred)
   report=classification_report(y_test,y_pred)
   mat=confusion_matrix(y_test,y_pred)
   dis=ConfusionMatrixDisplay(mat)    #object created to cnfusion matrix displaymat
   print(y_pred)
   print('*'*20)
   print(score)
   print('*'*20)
   print(mat)
   print('*'*20)
   print(report)
   dis.plot()
```
